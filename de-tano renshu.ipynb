```python
import os
import glob
import pandas as pd

%cd "C:\\Users\\kato\\OneDrive\\ドキュメント\\B2\\graduation research"

folder_path ="kasyo"

# フォルダ内の“kasyo”を含むファイル名を取得する
kasyo_files = glob.glob(os.path.join(folder_path, '*kasyo01*'))

kasyo_data = []
# 結果を表示します
for file in kasyo_files:
    dic_file = [file]
    print(file)
    kasyo_data = kasyo_data + dic_file
    
# 'kasyo_data'をpythonで読み込めるデータフレームにする
df = pd.read_excel(kasyo_data[0])
from IPython.display import display

header_list = df.columns.tolist()
list_length = len(header_list)
print(list_length)

col = 0
while col <= list_length - 1 :
    header_list[col]=col
    col=col+1

print(header_list)

df.columns = header_list

df = pd.DataFrame(df)
display(df)
```

```python
col1 = 75
col2 = 140
```

```python
# いらない(データが空白だったり、文字列だったりする)行を削除する
rows = df.shape[0]     # データが何行あるか(作業する幅)を'rows'として保存
del_rows = []          # 消したい行がどんどんぶち込まれていく空のゴミ箱(リスト)
now_row = 0            # 0行目からローラーしていく作戦

# 欠損値及び文字列を含む行を抽出（今欠損地があって困るのは75列目と140列目なので、75列目と140列目の0行目から順に欠損していないか確認していく）
while now_row < rows:
    cel1 = df.iloc[now_row, col1]        # 75列目、'now_row'行目のセルを'cel1'としてみる。
    cel2 = df.iloc[now_row, col2]       # 140列目、'now_row'行目のセルを'cel2'としてみる。

    # 'cel1'もしくは'cel2'が欠損値だった場合、その行をゴミ箱にぶち込む
    if True == pd.isna(cel1) or True == pd.isna(cel2):
        del_rows = del_rows + [now_row]
    # 'cel1'もしくは'cel2'が文字列だった場合もその行をゴミ箱にぶち込む
    elif isinstance(cel1,(str)) or isinstance(cel2,(str)):
        del_rows = del_rows + [now_row]
    # それ以外だった場合(基本的には数値しか残らない)、とりあえずやることなし
    else:
        pass
# 分岐してそれぞれの作業が終わったら、次の行へ移動
now_row = now_row+1


# 数値以外のcellを含む行を削除する（先ほどゴミ箱にぶち込んだ行を完全に削除する）
df_new = df.drop(del_rows)
df_new = df_new.reset_index(drop = True)    # 行のヘッダーがめちゃくちゃになったので整理

display(df_new)
```
